1. Problem Statement:
Find all the subsets from a set of numbers whose sum is zero.
Constraint: Subset size must be 5
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}


```python
from itertools import combinations

# Given set
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Constraint: Subset size = 5
result_5 = [subset for subset in combinations(numbers, 5) if sum(subset) == 0]

print("Subsets of size 5 with sum zero:")
for r in result_5:
    print(r)
```

Problem 2:
Find all subsets whose size is between 3 and 6 and whose sum is zero

```python
# Given set again
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Constraint: Subset size between 3 and 6
result_3_to_6 = []

for size in range(3, 7):
    result_3_to_6.extend([subset for subset in combinations(numbers, size) if sum(subset) == 0])

print("\nSubsets of size 3 to 6 with sum zero:")
for r in result_3_to_6:
    print(r)
```
